```python
class Binary:
    def __init__(self, user_list, user_find):
        if not isinstance(user_list, list):
            raise TypeError("Input must be a list")

        if not isinstance(user_find, int):
            raise TypeError("Target must be an integer")

        if user_list != sorted(user_list):
            raise ValueError("List must be sorted")

        self.lst = user_list
        self.target = user_find

    def binary_search(self):
        low = 0
        high = len(self.lst) - 1

        while low <= high:
            mid = (low + high) // 2

            if self.lst[mid] == self.target:
                return f"found at : {mid}"
            elif self.lst[mid] < self.target:
                low = mid + 1
            else:
                high = mid - 1

        return -1
if __name__ == "__main__":
    try:
        obj = Binary([1, 2, 3, 4, 5, 6], 4)
        print(obj.binary_search())

    except Exception as e:
        print(e)
```

```python
class Linear:
    def __init__(self, user_list, user_find):
        if not isinstance(user_list , list):
            raise TypeError("TypeError : the input is not list")

        if not isinstance(user_find , int):
            raise ValueError("ValueError : the input must be int")

        if user_list != sorted(user_list):
            raise ValueError("ValueError : input list should be sorted")
            
        self.list = user_list
        self.find = user_find

    def linear_serach(self):
        for index , value in enumerate(self.list):
            if value == self.find:
                return f"found at {index}"
        return -1

if __name__ == "__main__":
    try:
        user_obj = Linear([1,2,3,4,4,5,6,7] , 35)
        print(user_obj.linear_serach())

    except Exception as e:
        print(e)
```
