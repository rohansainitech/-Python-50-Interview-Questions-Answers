```python
def fib(n: int) -> list:
    """
    Generate Fibonacci series till the given number of terms.

    Args:
        n (int): Number of terms provided by user

    Returns:
        list: Fibonacci series
    """

    # Data Validation
    if not isinstance(n, int):
        raise TypeError("Input must be an integer")
    if n <= 0:
        raise ValueError("Number of terms must be positive")

    # Logic
    a, b = 0, 1
    series = []

    for _ in range(n):
        series.append(a)
        a, b = b, a + b

    return series


# -------- Main Execution --------
if __name__ == "__main__":
    try:
        num = int(input("Enter number of terms: "))
        result = fib(num)
        print("Fibonacci series is:", result)

    except (TypeError, ValueError) as e:
        print("Error:", e)
```
