```python
from typing import List, Any

def remove_duplicate(items: List[Any]) -> List[Any]:
    """
    Remove duplicates from a list while maintaining original order.
    """
    if not isinstance(items, list):
        raise TypeError("Input must be a list.")

    if len(items) == 0:
        raise ValueError("List must not be empty.")

    result = []

    for i in items:
        if i not in result:
            result.append(i)

    return result


# --------- main execution ----------
if __name__ == "__main__":
    try:
        # taking list input from user
        number_list = [int(x) for x in input("Enter numbers separated by space: ").split()]

        # calling the function
        result = remove_duplicate(number_list)

        # display the result
        print(f"After Removing Duplicates: {result}")

    except ValueError:
        print("Please enter only integers separated by space.")
    except TypeError as e:
        print(e)
```
