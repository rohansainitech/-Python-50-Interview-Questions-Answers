```python
# 15. Reverse_Words_in_a_Sentence

string = "P is fun"
reverse_sentence = string.split()
result = " ".join(reverse_sentence[::-1])
print(result)
```

```python
def reverse_sentence(string: str) -> str:
    """
    Reverse the sentence you write.
    
    Args:
        string (str): Sentence input by user
        
    Returns:
        str: Reversed sentence
    """
    
    # Data Validation
    if not isinstance(string, str):
        raise TypeError("Input must be a string")
        
    words = string.split()
    result = " ".join(words[::-1])
    return result


# -------- Main Execution --------
if __name__ == "__main__":
    try:
        user_input = input("Enter sentence to reverse: ")
        result = reverse_sentence(user_input)
        print(f"Reverse Sentence is: {result}")
    except TypeError as e:
        print("Error:", e)
```
