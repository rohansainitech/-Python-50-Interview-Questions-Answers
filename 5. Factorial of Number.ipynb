```python
def factorial(num: int)-> int:
    """
    Calculate the factorial of a non-negative integer.
    
    Parameter:
        num(int): The number for which factorial is required.
    Returns:
        int: Factorial of given number.
    Raises:
        ValueError: if num is negative>
        TypeError: if num is not an integer.
    """
    if isinstance(num,bool):
        raise TypeError("Boolean Values are not allowed")
    # check type of num
    if not isinstance(num,int):
        raise TypeError("Input must be an integer")
    # negetive check & Zero check
#     if num = 0:
#         raise ValueError("Factorial of zero is not define")
    if num < 0:
        raise ValueError("Factorial is not defined for negetive number")
    factorial_number = 1
    for i in range(number,0,-1):
        factorial_number = i * factorial_number
    return factorial_number

# main Execution
if __name__ == "__main__":
    try:
        number = input("Enter your number to find factorial: ")
        number = int(number)
        print("factorial :",factorial(number))
    except ValueError as e:
        print("Input must be a positive number")
    except TypeError as e:
        print("Error :",e)
```
