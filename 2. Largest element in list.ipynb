```python
from typing import List
def find_largest_number(number_list: List[float]) -> float:
    
    """
    Function to find largest number in the given list
    
    Args:
        number_list(number_list: list[float]): A list containing numeric value (int or float)
        
    Return: 
        TypeError: if input is not a list or contains non-numeric value
        ValueError: if the list is Empty
    """
    # Validate the input is a list
    if not isinstance(number_list,list):
        raise TypeError("Input must be a list of number float or int")
    # Validate that list is not Empty
    if not number_list:
        raise ValueError("List can not be empty")
    # Validate that list element should be numeric
    for num in number_list:
        if not isinstance(num,(int,float)):
            raise TypeError("All element must be numeric")
    largest = number_list[0]
    for num in number_list:
        if num > largest:
            largest = num
    return largest
# main execution
if __name__ == "__main__":
    try:
        # taking input from user
        user_input = input("Enter number in the list seperated by space int or float only: ")
        # converting user_input into list
        number_list = [float(item) for item in user_input.split()]
        # calling the function
        result = find_largest_number(number_list)
        #display the result 
        print("Largest number in the list is : ",result)
    except (TypeError,ValueError) as error:
        print(error)
    except Exception as e:
        print(e)
```
