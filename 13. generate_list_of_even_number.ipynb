```python
even_numbers = []

limit = int(input("Enter a number till you want a list: "))

for i in range(limit):
    if i % 2 == 0:
            even_numbers.append(i)

print(even_numbers)
```

```python
def generate_even_numbers(limit: int) -> list:
    """
    Generate a list of even numbers from 0 up to the given limit (exclusive).

    Args:
        limit (int): Upper limit for generating even numbers (must be non-negative).

    Returns:
        list: A list containing even numbers up to the given limit.

    Raises:
        TypeError: If the input is not an integer.
        ValueError: If the input is negative.
    """

    # Input validation
    if not isinstance(limit, int):
        raise TypeError("Input must be an integer")

    if limit < 0:
        raise ValueError("Input must be a non-negative integer")

    even_numbers = []

    for i in range(limit):
        if i % 2 == 0:
            even_numbers.append(i)

    return even_numbers


# -------- main execution --------
if __name__ == "__main__":
    try:
        num = int(input("Enter a number till you want the even number list: "))
        result = generate_even_numbers(num)
        print("Even numbers:", result)

    except ValueError as ve:
        print("Value Error:", ve)

    except TypeError as te:
        print("Type Error:", te)

    except Exception as e:
        print("Unexpected Error:", e)
```
