```python
def decimal_to_binary(a: int) -> str:
    """
    Convert a decimal integer into its binary representation.

    Args:
        a (int): Decimal integer

    Returns:
        str: Binary representation of the decimal number
    """

    # Data Validation
    if not isinstance(a, int):
        raise TypeError("Input must be an integer")
    if a < 0:
        raise ValueError("Only non-negative integers are allowed")

    # Edge case
    if a == 0:
        return "0"

    # Logic
    binary = ""
    decimal = a

    while decimal > 0:
        binary = str(decimal % 2) + binary
        decimal //= 2

    return binary


# -------- Main Execution --------
if __name__ == "__main__":
    try:
        result = decimal_to_binary(48)
        print(f"Binary value is: {result}")
    except (TypeError, ValueError) as e:
        print("Error:", e)
```
