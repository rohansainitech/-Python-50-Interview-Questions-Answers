```python
def reverse_string(input_string: str) -> str:
    """
    this is a function use to reverse a given string.
    Args:
        input_str(str): the input we provide that should be str
    Return:
        str: Return value we get that will be a string 
    Raises:
        TypeError: IF INPUT IS NOT A STRING 
    """
    #validate the input type
    if not isinstance(input_string,str):
        raise TypeError("Input must be a string")
    revesed_str = ""
    for char in input_string:
        revesed_str = char + revesed_str
    return revesed_str
#----------Main Execuation-----
if __name__ == "__main__":
    try:
        #Taking input from the user 
        user_input = input("Enter your string to reverse : ")
        #calling the function
        result = reverse_string(user_input)
        #Display the reverse string
        print(f"Reversed string : {result}")
    except TypeError as e:
        print(e)
    except Exception as e:
        print(e)
```
