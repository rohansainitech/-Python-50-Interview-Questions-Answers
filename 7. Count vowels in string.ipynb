```python
def given_string(string: str) -> int:
    """
    This function finds the number of vowels in a given string.
    """
    if not isinstance(string, str):
        raise TypeError("Input must be a string")

    if not string.isalpha():
        raise ValueError("String should contain only alphabets")

    string = string.lower()
    count = (string.count("a") + string.count("e") +
             string.count("i") + string.count("o") +
             string.count("u"))
    return count


# -------Main Execution-------
if __name__ == "__main__":
    try:
        string = input("Enter a string: ")
        result = given_string(string)
        print(f"Number of vowels are: {result}")
    except (TypeError, ValueError) as e:
        print(e)
```
