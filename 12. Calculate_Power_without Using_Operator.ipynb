```python
def power(x: int, y: int) -> int:
    """
    Calculate x raised to the power y without using the ** operator.

    Args:
        x (int): Base value
        y (int): Exponent (must be a non-negative integer)

    Returns:
        int: Result of x raised to the power y

    Raises:
        TypeError: If inputs are not integers
        ValueError: If exponent is negative
    """

    # Input validation
    if not isinstance(x, int) or not isinstance(y, int):
        raise TypeError("Both x and y must be integers")

    if y < 0:
        raise ValueError("Exponent must be a non-negative integer")

    result = 1

    for _ in range(y):
        result *= x

    return result


# -------- main execution --------
if __name__ == "__main__":
    try:
        x = int(input("Enter base (x): "))
        y = int(input("Enter exponent (y): "))

        print("Result:", power(x, y))

    except ValueError as ve:
        print("Value Error:", ve)

    except TypeError as te:
        print("Type Error:", te)

    except Exception as e:
        print("Unexpected Error:", e)
```
