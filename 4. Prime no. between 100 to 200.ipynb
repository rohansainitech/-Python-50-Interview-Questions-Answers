```python
def is_prime_number(number: int)-> bool:
    """
    Functionto check wheather a number is prime or not.
    
    Args:
        number (int): The number to check.
        
    Returns:
        bool: True if the number is prime, False otherwise.
        
    Raises:
        TypeError: if input is not an integer.
        ValueError: if number is less than 2 (since prime are >=2).
    """
    # Validate input an integer
    if not isinstance(number,int):
        raise TypeError("input must be an integer")
    # Number should be at least 2
    if number < 2:
        raise ValueError("Prime number are greater than or equal to 2")
    for i in range(2,int(number**.5)+1):
    
        if number % i == 0:
            return False
            break
    else:
        return True

if __name__ == "__main__":
    try:
        prime_number = []
        lower_limit = input("enter lower limit: ")
        upper_limit = input("enter upper limit: ")
        if not lower_limit.strip().isdigit():
            raise TypeError("Lower limit must be a valid integer (no decimal or letters).")
        if not upper_limit.strip().isdigit():
            raise TypeError("Upper limit must be a valid integer (no decimal or letters).")
        lower_limit = int(lower_limit)
        upper_limit = int(upper_limit)
        
        for i in range(lower_limit,upper_limit):
            if is_prime_number(i):
                print(f"{i} is a prime number")
                prime_number.append(i)
        print(prime_number)
    except (TypeError,ValueError) as e:
        print(e)
    except Exception as e:
        print(e)
```
