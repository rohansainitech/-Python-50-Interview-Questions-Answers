```python
def sum_of_digit(num: int)-> int:
    """
    Calculate the sum of digit of a non-negative integer.
    
    Parameter:
        num(int): The numbeer whose digits will be summed.
    
    Return:
        int: Sum of the digits.
        
    Raises: 
        TypeError: if input is not an integer.
        ValueError: if input is negative.
    """
    if not isinstance(num, int):
        raise TypeError("plz enter a integer value")
    if num < 0:
        raise ValueError("input must a positive number.")
    
    # Logic
    temp = 0
    while num > 0:
        temp = temp + (num%10)
        num = num // 10   
    return temp

# --------main Execution-------
if __name__ == "__main__":
    try:
        num = int(input("enter Your Number : "))
        result = sum_of_digit(num)
        print(f"Sum of digits: {result}")
    except (TypeError,ValueError) as error:
        print
    except Exception as e:
        print(e)
```
