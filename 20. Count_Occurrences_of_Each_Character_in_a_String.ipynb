```python
def char(text: str) -> int:
    """
    To find the number of characters in a string (excluding spaces).
    """

    # Data validation
    if not isinstance(text, str):
        raise TypeError("Input must be a string")

    # Logic
    count = 0
    for ch in text:
        if ch != " ":
            count += 1

    return count


# --------- Main Execution -------
if __name__ == "__main__":
    try:
        text = "srila prabhupada"
        result = char(text)
        print("Number of characters in your given string are:", result)

    except TypeError as e:
        print("Error:", e)
```
