```python
def intersection(list1: list, list2: list)-> list:
    """
    To find out the intersection value between two list
    
    Arg(list): given value that we provide should be list 
    
    Return: what we get in result that is also a list.
    """
    
    # data validation 
    if not isinstance(list1, list) or not isinstance(list2, list):
        raise TypeError("Enter a valid list")

        
    #logic 
    list3 = []
    for i in list1: 
        if i in list2: 
            if i not in list3: 
                list3.append(i)
    return list3

# -------main Execution------
if __name__ == "__main__":
    try:
        list1 = [2,45,2,5,2,64,23,15,11,108]
        list2 = [1,5,23,3,9,11,108]
        result = intersection(list1,list2)
        print("intersection value from given list are: ",result)
        
    except TypeError as e:
        print("error",e)
```
