```python
def find_largest(numbers: list) -> int:
    """
    Find the largest element in a list using a for loop.

    Args:
        numbers (list): List of integers

    Returns:
        int: Largest element in the list
    """

    # Data Validation
    if not isinstance(numbers, list):
        raise TypeError("Input must be a list")
    if len(numbers) == 0:
        raise ValueError("List cannot be empty")

    # Logic
    largest = numbers[0]

    for num in numbers:
        if num > largest:
            largest = num

    return largest


# -------- Main Execution --------
if __name__ == "__main__":
    try:
        data = [10, 45, 23, 89, 5]
        result = find_largest(data)
        print("Largest element in the list is:", result)

    except (TypeError, ValueError) as e:
        print("Error:", e)
```
